## Task 1: Form a training and testing dataset to be used for this analysis.
Note: Remember that we shouldn't touch the testing dataset until we're completely done with our model to ensure it's an accurate representation of the generalizability.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_recall_curve, roc_curve, auc
from collections import Counter
import numpy as np
from math import exp
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
```

```python
def train_test_split(df, frac=0.2):
    
    # get random sample 
    test = df.sample(frac=frac, axis=0)

    # get everything but the test sample
    train = df.drop(index=test.index)

    return train, test
```

```python
nba = pd.read_csv('nba_logreg.csv')
nba.dropna(inplace=True)
print(nba.shape)
nba_train, nba_test = train_test_split(nba, frac=0.2)

print(nba_train.shape)
print(nba_test.shape)
```

## Task 2: Using 10-fold Cross-Validation, identify the best threshold for predicting 1 or 0 for your logistic regression model that predicts whether they will last 5 years or not given their FG% and Games Played (GP).
The easiest way to do this would probably be a double for-loop, where the outer loop iterates over a list of possible threshold values and the inner loop is your 10-fold CV. You can then plot the thresholds vs the average accuracy for each threshold.

```python
nba = nba[['GP', 'FG%', 'TARGET_5Yrs']]
```

```python
X = nba[['GP', 'FG%']]
y = nba['TARGET_5Yrs']
```

```python
logreg = LogisticRegression()
kf = KFold(n_splits=10, shuffle=True, random_state=1)
```

```python
thresholds = np.arange(0.1, 0.9, 0.01)
average_accuracies = []
```

```python
for threshold in thresholds:
    accuracies = []
    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        logreg.fit(X_train, y_train)
        
        y_prob = logreg.predict_proba(X_test)[:, 1]
        y_pred = (y_prob >= threshold).astype(int)
        
        accuracy = accuracy_score(y_test, y_pred)
        accuracies.append(accuracy)
    
    average_accuracy = np.mean(accuracies)
    average_accuracies.append(average_accuracy)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(thresholds, average_accuracies, marker='o')
plt.title('Threshold vs. Average Accuracy')
plt.xlabel('Threshold')
plt.ylabel('Average Accuracy')
plt.grid(True)
plt.show()
```

```python
best_threshold_index = np.argmax(average_accuracies)
best_threshold = thresholds[best_threshold_index]
print(f'The best threshold is {best_threshold} with an average accuracy of {average_accuracies[best_threshold_index]:.4f}')
```

## Task 3: Using n-fold CV, estimate the accuracy of your chosen threshold on unseen data.
This should be very similar to the previous task except your using the best threshold from the previous task and doing n-fold CV to get an estimate of the accuracy on unseen data (since each observation will be predicted on when it is not in the training fold).

```python
best_threshold = thresholds[best_threshold_index]
print(best_threshold)
```

```python
best_threshold = 0.54
accuracies_n_fold = []

n_splits = 5
kf = KFold(n_splits=n_splits, shuffle=True, random_state=1)
```

```python
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    logreg.fit(X_train, y_train)

    y_prob = logreg.predict_proba(X_test)[:, 1]
    y_pred = (y_prob >= best_threshold).astype(int)

    accuracy = accuracy_score(y_test, y_pred)
    accuracies_n_fold.append(accuracy)

mean_accuracy_n_fold = np.mean(accuracies_n_fold)

print(f'Estimated mean accuracy using best threshold with {n_splits}-fold CV: {mean_accuracy_n_fold:.4f}')
```

## Task 4: Using the bootstrap, get the mean and standard error of each coefficient for your logistic regression model.
Remember that the bootstrap just involves taking repeated samples from your original dataset up to the size of that dataset but with replacement. The number of samples you draw is your choice but somewhere around 11 is a fine number.

```python
coef_means = []
bootstrap_coefs = []
```

```python
n_bootstrap = 11
for i in range(n_bootstrap):
    bootstrap_sample = nba.sample(n=len(nba), replace=True)

    X_bootstrap = bootstrap_sample[['GP', 'FG%']]
    y_bootstrap = bootstrap_sample['TARGET_5Yrs']

    logreg.fit(X_bootstrap, y_bootstrap)

    bootstrap_coefs.append(logreg.coef_[0])

coef_means = np.mean(bootstrap_coefs, axis=0)
coef_std_errs = np.std(bootstrap_coefs, axis=0)

print(f'Coefficient for Games Played: Mean = {coef_means[0]:.4f}, Standard Error = {coef_std_errs[0]:.4f}')
print(f'Coefficient for Field Goal Percentage: Mean = {coef_means[1]:.4f}, Standard Error = {coef_std_errs[1]:.4f}')
```

## Task 5: Use your bootstrap estimates of your coefficients to predict the probability of lasting 5 years or not for your test data. If you use the threshold from before, what accuracy do you get? Is this similar to what you saw from n-fold CV?
Note: Look back at the Logistic Regression slides to see how to calculate probability.

```python
bootstrap_coefs = []
predicted_probs = []
```

```python
for _ in range(n_bootstrap):
    bootstrap_sample = nba.sample(n=len(nba), replace=True)

    X_bootstrap = bootstrap_sample[['GP', 'FG%']]
    y_bootstrap = bootstrap_sample['TARGET_5Yrs']

    logreg.fit(X_bootstrap, y_bootstrap)

    bootstrap_coefs.append(logreg.coef_[0])
    predicted_probs.append(logreg.predict_proba(nba_test[['GP', 'FG%']])[:, 1])


mean_probs = np.mean(predicted_probs, axis=0)


threshold = 0.54 # Best threshold from Task 2
predicted_classes = (mean_probs >= threshold).astype(int)
accuracy = accuracy_score(nba_test['TARGET_5Yrs'], predicted_classes)

print(f'Accuracy using bootstrap: {accuracy:.4f}')
```

Both accuracies are around 70%, likely because the thresholds are the same.
