# $K$-Nearest Neighbors and Decision Trees

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs, make_circles, make_classification, fetch_olivetti_faces
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from collections import Counter
from sklearn.metrics import confusion_matrix, accuracy_score, precision_recall_curve, roc_curve, auc
from sklearn.model_selection import train_test_split
import pandas as pd

%matplotlib inline
```

```python
def plot_decision_boundary(X, y, clf):
    x0_mesh, x1_mesh = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, 0.1),
                                 np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, 0.1))

    y_mesh_pred = clf.predict(np.c_[x0_mesh.ravel(), x1_mesh.ravel()])
    y_mesh_pred = y_mesh_pred.reshape(x0_mesh.shape)

    plt.contourf(x0_mesh, x1_mesh, y_mesh_pred, cmap="RdBu", vmin=0, vmax=1)
    plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, edgecolor = 'w', cmap="RdBu", vmin=-.2, vmax=1.2, linewidth=1)
    plt.show()
```

## Linear Decision Boundary with Logistic Regression

```python
X, y = make_classification(
    n_samples=1000, n_features=2, n_redundant=0, n_informative=2, random_state=10, n_clusters_per_class=1
)

x_train = X[:500,:]
x_test = X[500:,:]
y_train = y[:500]
y_test = y[500:]
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1], c = y_train, s = 50, edgecolor = 'w', cmap="RdBu", vmin=-.2, vmax=1.2, linewidth=1)
```

```python
clf = LogisticRegression().fit(x_train, y_train)
train_acc = clf.score(x_train,y_train)
test_acc = clf.score(x_test, y_test)

print(f'Training Accuracy: {train_acc}')
print(f'Test Accuracy: {test_acc}')
plot_decision_boundary(x_test,y_test,clf)
```

## Non-Linear Decision Boundary with Logistic Regression

```python
X, y = make_circles(n_samples=1000, noise=0.2, factor=0.01, random_state=1000)

x_train = X[:500,:]
x_val = X[500:600,:]
x_test = X[600:,:]
y_train = y[:500]
y_val = y[500:600]
y_test = y[600:]
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1], c = y_train, s = 50, edgecolor = 'w', cmap="RdBu", vmin=-.2, vmax=1.2, linewidth=1)
```

```python
clf = LogisticRegression().fit(x_train, y_train)
train_acc = clf.score(x_train,y_train)
test_acc = clf.score(x_test, y_test)

print(f'Training Accuracy: {train_acc}')
print(f'Test Accuracy: {test_acc}')
plot_decision_boundary(x_test,y_test,clf)
```

### Effect of varying $K$

```python
k_values = np.arange(1,20)
```

```python
train_acc_list = []
val_acc_list = []
for k in k_values:
    clf = KNeighborsClassifier(n_neighbors=k).fit(x_train, y_train)
    train_acc = clf.score(x_train,y_train)
    val_acc = clf.score(x_val, y_val)
    val_acc_list.append(val_acc)
    train_acc_list.append(train_acc)
    print('Number of neighbors: ' + str(k))
    plot_decision_boundary(x_val,y_val,clf)
```

```python
plt.plot(k_values,train_acc_list)
plt.plot(k_values,val_acc_list)
```

```python
clf = KNeighborsClassifier(n_neighbors=12).fit(x_train, y_train)
clf.score(x_test,y_test)
```

## Decision Trees
https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

```python
clf = DecisionTreeClassifier().fit(x_train, y_train)
train_acc = clf.score(x_train,y_train)
test_acc = clf.score(x_test, y_test)

print(f'Training Accuracy: {train_acc}')
print(f'Test Accuracy: {test_acc}')
plot_decision_boundary(x_test,y_test,clf)
```

### Cost Complexity Pruning

The most common method of pruning is referred to as cost complexity pruning (CCP), it also known as weakest link pruning. Rather than considering every possible subtree though, CCP seeks to minimize the following equation for all values of the tuning parameter $\alpha$

$$
\sum_{m=1}^{\lvert T \rvert} \sum_{i: x_i\in R_m}(y_i - \hat{y}_{R_m})^2 + \alpha \lvert T\rvert
$$

where $\lvert T\rvert$ indicates the number of terminal nodes. The effect of this is that for various sizes of subtrees we will be able to identify the optimal subtree.

```python
path = clf.cost_complexity_pruning_path(x_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

plt.plot(ccp_alphas, impurities)
plt.xlabel("alpha")
plt.ylabel("total impurity of leaves")
```

```python
clfs = []

for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(x_train, y_train)
    clfs.append(clf)
    
tree_depths = [clf.tree_.max_depth for clf in clfs]
plt.figure(figsize=(10,  6))
plt.plot(ccp_alphas[:-1], tree_depths[:-1])
plt.xlabel("effective alpha")
plt.ylabel("total depth")
```

```python
acc_scores = [clf.score(x_val, y_val) for clf in clfs]

tree_depths = [clf.tree_.max_depth for clf in clfs]
plt.grid()
plt.plot(ccp_alphas[:-1], acc_scores[:-1])
plt.xlabel("alpha")
plt.ylabel("Accuracy scores")
```

```python
optimal_clf = clfs[np.argmax(acc_scores)]

train_acc = optimal_clf.score(x_train,y_train)
test_acc = optimal_clf.score(x_test, y_test)

print(f'Training Accuracy: {train_acc}')
print(f'Test Accuracy: {test_acc}')

plot_decision_boundary(x_test,y_test,optimal_clf)
```

```python
plt.figure(figsize=(10, 10))
plot_tree(optimal_clf)
plt.show()
```

# Epileptic Seizure Recognition
https://www.kaggle.com/datasets/harunshimanto/epileptic-seizure-recognition

```python
epil = pd.read_csv('Epileptic Seizure Recognition.csv')
print(epil.shape)
epil.head()
```

```python
Counter(epil.y)
```

```python
epil['y'] = np.where(epil['y'] == 1,1,0)
```

```python
epil.columns
```

```python
X = epil.drop(columns = ['Unnamed','y'])
y = epil['y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```

```python
print(X_train.shape, X_val.shape, X_test.shape)
```

### KNN

```python
k_values = np.arange(1,20)

train_acc_list = []
val_acc_list = []

for k in k_values:
    clf = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    train_acc = clf.score(X_train,y_train)
    val_acc = clf.score(X_val, y_val)
    val_acc_list.append(val_acc)
    train_acc_list.append(train_acc)
    print('Number of neighbors: ' + str(k) + ' Val Acc: ' + str(val_acc))
```

```python
plt.plot(k_values, train_acc_list)
plt.plot(k_values, val_acc_list)
```

```python
clf = KNeighborsClassifier(n_neighbors=1).fit(X_train, y_train)
```

```python
preds = clf.predict(X_test)
```

```python
confusion_matrix(y_test, preds)
```

```python
clf.score(X_test,y_test)
```

### Decision Tree

```python
clf = DecisionTreeClassifier().fit(X_train, y_train)
train_acc = clf.score(X_train,y_train)
val_acc = clf.score(X_val, y_val)

print(f'Training Accuracy: {train_acc}')
print(f'Val Accuracy: {val_acc}')
```

```python
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

plt.plot(ccp_alphas, impurities)
plt.xlabel("alpha")
plt.ylabel("total impurity of leaves")
```

```python
len(ccp_alphas)
```

```python
clfs = []

i = 0
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)
    print(i)
    i+=1
    
tree_depths = [clf.tree_.max_depth for clf in clfs]
plt.figure(figsize=(10,  6))
plt.plot(ccp_alphas[:-1], tree_depths[:-1])
plt.xlabel("effective alpha")
plt.ylabel("total depth")
```

```python
acc_scores = [clf.score(X_val, y_val) for clf in clfs]

tree_depths = [clf.tree_.max_depth for clf in clfs]
plt.grid()
plt.plot(ccp_alphas[:-1], acc_scores[:-1])
plt.xlabel("alpha")
plt.ylabel("Accuracy scores")
```

```python
optimal_clf = clfs[np.argmax(acc_scores)]

train_acc = optimal_clf.score(X_train,y_train)
test_acc = optimal_clf.score(X_test, y_test)

print(f'Training Accuracy: {train_acc}')
print(f'Test Accuracy: {test_acc}')
```
