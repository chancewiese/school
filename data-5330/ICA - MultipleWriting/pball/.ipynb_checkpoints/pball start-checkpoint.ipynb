```python
import pandas as pd
import psycopg2

# BEFORE RUNNING, MAKE SURE THIS NOTEBOOK IS IN THE SAME DIRECTORY AS ALL THE .CSV FILES IT CAME WITH!

#establishing the connection
conn = psycopg2.connect(
    database="pball", 
    user='[YOUR AWS USERNAME]', 
    password='[YOUR AWS PASSWORD]', 
    host='[YOUR AWS ENDPOINT]', 
    port= '5432'
)

#Creating a cursor object using the cursor() method
cursor = conn.cursor()

query = '''CREATE TABLE IF NOT EXISTS result (
           resultid SERIAL PRIMARY KEY,
           homeid INT,
           awayid INT,
           venueid INT,
           hscore SMALLINT,
           ascore SMALLINT
        );'''

cursor.execute(query)
conn.commit()

query = '''CREATE TABLE IF NOT EXISTS venue (
           venueid SERIAL PRIMARY KEY,
           locname VARCHAR(100) NOT NULL,
           city VARCHAR(100)
        );'''

cursor.execute(query)
conn.commit()

query = '''CREATE TABLE IF NOT EXISTS player (
           pnum INT PRIMARY KEY,
           first VARCHAR(80),
           last VARCHAR(80) NOT NULL,
           level SMALLINT
        );'''

cursor.execute(query)
conn.commit()
```

```python
import pandas as pd

df = pd.read_csv('player.csv')

for x in df.index:
    query = """INSERT INTO player (pnum, first, last, level)
    VALUES (%s, '%s', '%s', %s)""" % (df['pnum'].loc[x], df['first'].loc[x],
                                     df['last'].loc[x], df['level'].loc[x])
    
    cursor.execute(query)
    conn.commit()
    
df = pd.read_csv('venue.csv')

for x in df.index:
    query = """INSERT INTO venue (locname, city)
    VALUES ('%s', '%s')""" % (df['LocName'].loc[x], df['City'].loc[x])
    
    cursor.execute(query)
    conn.commit()
    
df = pd.read_csv('result.csv')

for x in df.index:
    query = """INSERT INTO result (homeid, hscore, awayid, ascore, venueid)
    VALUES (%s, %s, %s, %s, %s)""" % (df['Home'].loc[x], df['Hscore'].loc[x],
                                     df['Away'].loc[x], df['Ascore'].loc[x],
                                     df['Venue'].loc[x])
    
    cursor.execute(query)
    conn.commit()
    
```
